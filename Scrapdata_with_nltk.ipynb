```
import requests
```

```
response=requests.get('https://en.wikipedia.org/wiki/A._P._J._Abdul_Kalam')
```

```
response
```

```
webdata=response.text
```

```
webdata
```

```
from bs4 import BeautifulSoup
```

```
soup=BeautifulSoup(webdata,'html5lib')
```

```
data=soup.get_text()
```

```
with open('data.txt','w+') as f:
  f.write(data)
```

```
f=open('data.txt','r')
x=f.read()
f.close()
```

```
newdata=[i for i in x.split()]
```

```
newdata
```

```
len(newdata)
```

```
import nltk
```

```
freq=nltk.FreqDist(newdata)
```

```
freq.plot(20)
```

# Remove stopwords

```
from nltk.corpus import stopwords
```

```
nltk.download('stopwords')
```

```
print(stopwords.words('english'))
```

```
remove_data=[i for i in newdata if i.lower() not in stopwords.words('english')]
```

```
len(remove_data)
```

```
freq1=nltk.FreqDist(remove_data)
```

```
freq1.plot(20)
```
